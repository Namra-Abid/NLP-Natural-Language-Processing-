```python
import re
import nltk
from nltk.corpus import stopwords

from nltk.stem import WordNetLemmatizer
```

```python
paragraph = """Introduce yourself, whether in a professional capacity or a casual one, is vital to how people will perceive you for the following interactions that take place. It is essential to create a near-accurate image of yourself in their mind that they feel comfortable and courteous while interacting with you and yet considering that you know what you do.

Whether to say “Hi, I am XYZ” and scratch your head, or say I am this and that is no confident indication of who you are. The right introduction is not too long or too short. It does not have to be a crammed speech. Your introduction does not have to be completely technical or professional."""
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
sentences=nltk.sent_tokenize(paragraph)
for i in range(len(sentences)):
    remove_punc=re.sub('[^a-zA-Z]',' ',sentences[i])
    lower_sentence=remove_punc.lower()
    lower_sentence=lower_sentence.split()
    sentence_clean=[]
    for word in lower_sentence:
        if word not in set(stopwords.words('english')):
            sentence_clean.append(lemmatizer.lemmatize(word))
    review=' '.join(sentence_clean)
    #print(review)
    corpus.append(review)
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
X = cv.fit_transform(corpus).toarray()
```
